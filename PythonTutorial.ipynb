## Exercise 2: Fourier Transforms

```python
Fs = 1000  #Sampling frequency (number of pts per second)
dt = 1./Fs #Note what would happen if this were "dt = 1/Fs"
t = np.arange(0,10,dt) #Creates a 1D array from 0 to 10 in steps of dt
y = 3*np.sin(2*np.pi*10*t)
plt.plot(t[:500], y[:500]) #Only plotting the first 500 points
```

```python
Y = np.fft.fft(y)
freq = np.fft.fftfreq(len(y), dt)
plt.plot(freq, abs(Y))
```

```python
plt.semilogy(freq[:len(freq)/2], abs(Y)[:len(freq)/2])
```
